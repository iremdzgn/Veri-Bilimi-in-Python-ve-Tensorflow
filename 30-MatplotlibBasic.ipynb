```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasList = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloList =[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasList = np.array(yasList)
numpyKiloList = np.array(kiloList)
```

```python
numpyYasList
```

```python
numpyKiloList
```

```python
plt.plot(numpyYasList,numpyKiloList,"g")  #yas listesi x ekseninde kilo y ekseninde yazılır ve "g" harfi rengi ifade eder
```

```python
plt.plot(numpyKiloList,numpyYasList,"y") #x ekseninde kilo y ekseninde yas
```

```python
plt.plot(numpyYasList,numpyKiloList,"b")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yasa Göre Degisim Grafigi")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)#1 sıra olacak ,2 kolon olacak,şuan 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.4,0.4,0.6,0.6])  #1. ve 2. x ve y eksenlerinin hangi değerden başlıyacağı 3. weight 4. height göstermektedir
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik başlıgı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
(myFigure,myEksenler) = plt.subplots(nrows=1,ncols=2)
type(myEksenler)

for eksen in myEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
    
plt.tight_layout() #arasını açıp düzgün göstermek için yapılır
```

```python
yeniFigure = plt.figure(figsize= (3,3))

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label = "numpyDizisi ** 3")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**4,label = "numpyDizisi **4")
yeniEksen.legend(loc=2) #loc=2 legend'ın nerede çıkacağını ayarlamaya yarar. loc=2 yapmazsak eğer varsayım olarak loc=2 de gösterilir çeşitli sayılar vererek yerini değiştirebiliriz
```
