```python
dataFrame = dataFrame.drop("transmission", axis = 1) # transmission kolonunda sayısal bir değer olmadığı için regresyona soktuğumuzda hata vereceği için o kolonu çıkarırız
```

```python
y = dataFrame["price"].values #ulaşmak istediğimiz şey
x = dataFrame.drop("price", axis = 1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #modeli oluşturmak için
from tensorflow.keras.layers import Dense #katmanları oluşturmak için
```

```python
x_train.shape #5 özlellik ve 9090 veri var
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1)) # çıkış katmanı

model.compile(optimizer = "adam" , loss="mse")
```

```python
model.fit(x = x_train, y = y_train, validation_data=(x_test,y_test), batch_size=250, epochs=300)  # validation data (doğrulama verisini) direkt modele verebiliriz
```

```python
kayipVeri = pd.DataFrame(model.history.history)
```

```python
kayipVeri.head()
```

```python
kayipVeri.plot() #azalan bir grafik ve loss ve validation loss birlikte hareket ediyor birbirinden çok bağımsız değiller.eğer iki çizginin arasında boşluk çok olursa overfitting yaptığımız anlamına gelir
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi) ##y_test ile tahmin dizisinin hata kıyaslamasında 3176 poundluk bir fark vardır
```
