```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **2
```

```python
(benimFigure, benimEksen) =plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color="#99004C",alpha=0.5) #alpha transparanlık katar 1 opaktır birden aşağıya indikçe saydamlaşır
benimEksen.plot(numpyDizisi2,numpyDizisi1, color = "#21971C")
```

```python
(yeniFigure,yenieksen) = plt.subplots()
yenieksen.plot (numpyDizisi1,numpyDizisi1 +2,color="#99004C" , linewidth = 6.0)
yenieksen.plot(numpyDizisi1,numpyDizisi1+3,color="#1E90FF", linewidth=4.0)

yenieksen.plot(numpyDizisi1,numpyDizisi1+4, color="#9914CA",linewidth=2.0,linestyle="--")
yenieksen.plot(numpyDizisi1,numpyDizisi1+5, color="#38BDE5", linewidth=2.0,linestyle=":")
yenieksen.plot(numpyDizisi1,numpyDizisi1+6, color="#FF1493",linewidth=2.0,linestyle ="-.")

yenieksen.plot(numpyDizisi1, numpyDizisi1+7,  color="#32CD32", linewidth=2.0, linestyle = "--", marker = "o", markersize = 6, markerfacecolor = "red")
yenieksen.plot(numpyDizisi1, numpyDizisi1+8,  color="#FA8072", linewidth=2.0, marker = "*", markersize = 8)
```
